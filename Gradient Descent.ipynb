### General Function
Here I have coded a general function to solve for minimum with gradient descent with multiple variables.
As these may contain more than 2 variable, it will not be possible to plot it, so I have only solved it numerically, the path is there but just not visualised.
The parameters to the functions are: 

`f` which is the function itself

`list_derivative` which contains a list of names of the derivative functions (provided they take all the variables as parameters)

`list_start` which has a list of starting coordinates

`lr` which defines the learning rate

`itr` which has the number of iterations to be run through.

The final minimised value is then printed.

Below I have used the functions we have defined above and as we can see the results match.

```python
def multivar(f,list_derivative,list_start,lr,itr):
    num=len(list_start)
    best=list_start
    temp=[0 for i in range(num)]
    all_coord = [[] for i in range(itr)]
    final=[]
    for i in range(itr):
        for j in range(num):
            temp[j] = best[j] - list_derivative[j](*best)*lr
            best[j]= temp[j]
            all_coord[i].append(temp[j])
        final.append(f(*all_coord[i]))
    print("The minimum value of the function for given learning rate and starting point with gradient descent is :",final[-1])   
```

```python
multivar(f4,[df4_dx,df4_dy],[1,0],0.05,1000)
```

```python
multivar(f5,[df5_dx],[3.1],0.05,500)
```
